```python
import speech_recognition as sr     # it is use to recognize our voice
import pyttsx3       # it help in speak to us 
import pywhatkit
import datetime
import wikipedia
import pyjokes



speak = sr.Recognizer()
assistant = pyttsx3.init()     # it start the microphone o speak
voices = assistant.getProperty('voices')
assistant.setProperty('voice',voices[1].id)
assistant.say("hello i am your assistnat")
assistant.say("waht can i help you")
assistant.runAndWait()


def talk(text): # it repeat the text what we speak
    assistant.say(text)
    assistant.runAndWait()

def take_command():
    try:
        with sr.Microphone() as source:          # this is use for using the microphone as an input 
            print(".....please speak.....")
            voice = speak.listen(source)            # it will call the recognizer to listen this source
            Command = speak.recognize_google(voice)  # it will help us to convert voice into text 
            talk(Command)
            return Command
    except:
        pass
   
 
       
    
def run_assistant():      # it is use to play a song
    Command = take_command()      # call uper function
    print(Command)

    if 'play'in Command:
        sing = Command.replace('play','')     # replace play from voice into noting 
        talk('playing'+ sing ) 
        pywhatkit.playonyt(sing) 
        print(sing) 
        
    elif 'time'in Command:
        clock = datetime.datetime.now().strftime("%I:%M %p")
        talk("the time is" + clock)
        print(clock)
    elif 'search for'in Command:
        person = Command.replace('search for','')
        info = wikipedia.summary(person,1)     # summary is ued for search in google and 1 stand for only one line
        talk(info)
        print(info)
    elif 'joke'in Command:
        talk(pyjokes.get_joke())
        print(pyjokes.get_joke())
    elif 'are you single'in Command:
        talk('no i have a relation with the computer')
    elif 'what are you doing'in Command:
        talk('i am just talking with you')
    elif 'can we go'in Command:
        talk('no i have some work')
    else :
        if 'bye bye assistant'in Command:
            print("bye bye")
       

    
while(True):
    
     run_assistant()
     
       
   
```
